```python
players = pd.read_csv("players_data.csv")
```

```python
players.info()
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
players.tail()
```

```python
players.shape[0]
```

```python
players.groupby(["Player", "Year"]).get_group('')
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```
